```python
from os import listdir
from os.path import isfile, join


dirpath='./abbasi/'
onlyfiles = [f for f in listdir(dirpath) if isfile(join(dirpath, f))]

onlyfiles[0].split('.')[0]
ps=[]

for f in onlyfiles:
    name=f.split('.')[0]
    fullpath=dirpath+f
    df = pd.read_json(codecs.open(fullpath, 'r', 'utf-8'))
    for i in range(len(df)):
        for g in range(len(df.poets[i]['homes'])):
            d={'name':name,'poet':df.poets[i]['title'],'0':df.poets[i]['homes'][g][0],'1':df.poets[i]['homes'][g][1]}
            ps.append(d)
refined=pd.DataFrame(ps)
refined.head()
```
